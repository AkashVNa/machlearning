outliers are present in almost every features except 'worst concave points' hence, outliers need to be removed.
Outliers are mostly seen beyond the upper fence and so only those are to be removed

```python
#functon to remove outliers
def removeoutlier(dat,column):
    d=dat.copy()
    for i in column:
        iqr=d[i].quantile(0.75)-d[i].quantile(0.25)
        d.loc[d[i]>d[i].quantile(0.75)+1*iqr,i]=d[i].median()
    return d

column=dat.columns.tolist()
column.remove('worst concave points')
new_data=removeoutlier(dat,column)
```

```python
new_data.skew().sort_values(ascending=False)
#to prevent over correction of data, this level of skewness is accepted
```

No object features are present, all are numericals hence no encoding required

```python
#scaling of data
from sklearn.preprocessing import StandardScaler
stdscale=StandardScaler()
X=stdscale.fit_transform(dat)
y=data.target
```

```python
#splitting data to determine the accuracy
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, random_state=188)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

Logistic Regression

Logistic Regression is mainly used for classification problems and uses the conditions of the independent variables(features) to find the probability of the dependent variable(target) belonging to a class. Since the problem under consideration is of classification type and the target variable (dependent variable) is a binomial variable(either 0 or 1) this method can be used. 

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
print("LOGISTIC REGRESSION")
print("Accuracy score: ", accuracy_score(y_test,y_pred))
print("Classification report is: ")
print(classification_report(y_test,y_pred))
logacc=accuracy_score(y_test,y_pred)
print("Heat map of confusion matrix is :")
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
plt.xlabel('predicted values')
plt.ylabel('actual values')
plt.show()
```

Decision Tree Classifier

Decision tree classifier is used mainly for classification problems and consist of tree like structures for evaluation starting from a root node and then through decision nodes and ending in leaf node. By incorporating necessary decision evaluations, the problem can be classified into specific leaf nodes by taking the appropriate node path. Since the problem under consideration is of classification type and the target variable (dependent variable) is a binomial variable(either 0 or 1) this method can be used.

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
print("DECISION TREE CLASSIFIER")
print("Accuracy score: ", accuracy_score(y_test,y_pred))
print("Classification report is: ")
print(classification_report(y_test,y_pred))
treeacc=accuracy_score(y_test,y_pred)
print("Heat map of confusion matrix is :")
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
plt.xlabel('predicted values')
plt.ylabel('actual values')
plt.show()
```

Random Forest Classifier

Random Forest uses multiple decision trees to reach a single result. This is also used mainly for classification problems. Two types of methods are used ie bagging and boosting. In bagging, random data is selected from the training set with repetition allowed and different models are created independently and majority of these models will give a better result than a single decision tree. In boosting, the decision trees are arranged in sequential order and the errors in preceeding decision tree is corrected in the succeeding decision tree so the final output from the last decision tree will be supposedly free of error. Since the problem under consideration is of classification type and the target variable (dependent variable) is a binomial variable(either 0 or 1) this method can be used.

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
print("RANDOM FOREST CLASSIFIER")
print("Accuracy score: ", accuracy_score(y_test,y_pred))
print("Classification report is: ")
print(classification_report(y_test,y_pred))
foracc=accuracy_score(y_test,y_pred)
print("Heat map of confusion matrix is :")
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
plt.xlabel('predicted values')
plt.ylabel('actual values')
plt.show()
```

Support Vector Machine

Support vector machine classifier tries to identify an optimal hyper plane which exists in an n-dimensional space(n=no of features or independent variables) that seperates the data points of respective classes. The hyperplane is selected in such a way that the distance of the closest point of the classes from the hyperplane is maximum. Since the problem under consideration is of classification type and the target variable (dependent variable) is a binomial variable(either 0 or 1) this method can be used.

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print("SUPPORT VECTOR CLASSIFIER")
print("Accuracy score: ", accuracy_score(y_test,y_pred))
print("Classification report is: ")
print(classification_report(y_test,y_pred))
svcacc=accuracy_score(y_test,y_pred)
print("Heat map of confusion matrix is :")
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
plt.xlabel('predicted values')
plt.ylabel('actual values')
plt.show()
```

K-Nearest Neighbors

K-Nearest Neighbor classification is based on the theory that data points of classes lies in close proximity to each other. To classify an unknown datapoint, the proximities with known data points are calculated and by selecting the appropriate weightage, the unknown point can be considered to be belonging to a specific class. Since the problem under consideration is of classification type and the target variable (dependent variable) is a binomial variable(either 0 or 1) this method can be used.

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
knc.fit(X_train,y_train)
y_pred=knc.predict(X_test)
print("K-NEAREST NEIGHBOR")
print("Accuracy score: ", accuracy_score(y_test,y_pred))
print("Classification report is: ")
print(classification_report(y_test,y_pred))
knnacc=accuracy_score(y_test,y_pred)
print("Heat map of confusion matrix is :")
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
plt.xlabel('predicted values')
plt.ylabel('actual values')
plt.show()
```

```python
com=pd.DataFrame([logacc,treeacc,foracc,svcacc,knnacc],index=['logistic regression','decision tree classifier','random forest classifier','support vector machine','k-nearest neighbors'],columns=['accuracy score'])
com
```
